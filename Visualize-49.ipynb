```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import os

# Load dataset
with open("/users/jboskovi/sionna_project/sionna_osm_scene/dataset.json") as f:
    dataset = json.load(f)

print(f"Total samples: {len(dataset)}")

# Output directory for verification images
output_dir = "verification_images"
os.makedirs(output_dir, exist_ok=True)

for idx, sample in enumerate(dataset):
    sensor_pos = sample["sensor_pos"]
    coverage_map = np.array(sample["coverage_map"])

    fig, ax = plt.subplots(figsize=(8, 12))
    ax.imshow(coverage_map, origin='lower', cmap='viridis',
              extent=[-105, 105, -164, 164])

    ax.plot(sensor_pos[0], sensor_pos[1], 'r*', markersize=15)

    # Coverage legend
    legend_elements = [
        Patch(facecolor='yellow', label='Covered (RSS > -90 dBm)'),
        Patch(facecolor='purple', label='Not covered (RSS ≤ -90 dBm)')
    ]
    ax.legend(handles=legend_elements, loc='lower right')

    n_covered = coverage_map.sum()
    ax.set_title(f"Sample {idx} | Sensor: {sensor_pos}\nCoverage: {n_covered}/{coverage_map.size} cells ({100*n_covered/coverage_map.size:.1f}%)")
    ax.set_xlabel("X (m)")
    ax.set_ylabel("Y (m)")

    plt.tight_layout()
    plt.savefig(f"{output_dir}/sample_{idx:03d}.png", dpi=80)
    plt.close()
    print(f"Saved sample_{idx:03d}.png — sensor at {sensor_pos}")

print(f"\nDone. {len(dataset)} images saved to '{output_dir}/'")
```
